```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = np.poly1d([-1/72**2, 0, 1])
```

```python
x = np.linspace(0, 72, 100)
plt.plot(x, y(x))
plt.plot([24, 24], [0, y(24)], color='black')
plt.plot([48, 48], [0, y(48)], color='black')
plt.plot([24, 48], [y(24), y(48)], 'o')
plt.annotate('  {0:.2f}'.format(y(24)), xy=(24, y(24)))
plt.annotate('  {0:.2f}'.format(y(48)), xy=(48, y(48)))
plt.ylim(0, plt.ylim()[1])
plt.xticks([0, 12, 24, 36, 48, 60, 72])
plt.xlim(0, plt.xlim()[1])
plt.ylabel('Percent of total points')
_ = plt.xlabel('Hours late')
```
