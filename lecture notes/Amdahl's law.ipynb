```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def amdahl(f, p):
    """f is the fraction parallelizable of the application.
       p is the number of cores.
       This function returns the speedup.
    """
    return 1/(1-f+f/p)
def plotAmdahl(f, mx=64):
    x = np.linspace(1,mx)
    plt.plot(x, [amdahl(f, p) for p in x], label="{:.1f}% fraction parallel".format(f*100))
```

```python
plt.figure()
plotAmdahl(0.5)
plotAmdahl(0.9)
plotAmdahl(0.99)
plt.legend(loc='upper left')
plt.xlabel('Number of cores')
plt.ylabel('Speedup')
plt.tight_layout()
```

```python
plt.figure()
plotAmdahl(0.5, 1024)
plotAmdahl(0.9, 1024)
plotAmdahl(0.99, 1024)
plt.legend(loc='center right')
plt.xlabel('Number of cores')
plt.ylabel('Speedup')
plt.tight_layout()
```

```python
plt.figure()
plotAmdahl(0.5, 256)
plotAmdahl(0.9, 256)
plotAmdahl(0.99, 256)
plt.legend(loc='center right')
plt.xlabel('Number of cores')
plt.ylabel('Speedup')
plt.tight_layout()
```
